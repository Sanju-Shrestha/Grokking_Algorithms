The binary_search function takes a **sorted array** and an item. If the item is in the array, the function returns its position. You’ll keep track of what part of the array you have to search through. At the beginning, this is the entire array:




```python
# Binary Search Function

def binary_search(user_list: list,item: int) -> int:
    # Low and high keep track of which part of 
    # the list you’ll search in.
    low = 0
    high = len(user_list) - 1
    while low <= high:
        mid = (low + high)//2
        guess = user_list[mid]
        # If the item is found
        if guess == item:
            return mid
        # If item is less than the guess
        if guess > item:
            high = mid - 1
        # When item is greater than the guess
        else:
            low = mid + 1
        # When the item is not in the list
    return None
```

```python
new_list = [10,20,30,80,115,222,333,1000]

# Function call for item 222
print('The item is at the index: ', binary_search(new_list,222))

# Function call for item 1001
print('The item is at the index: ', binary_search(new_list,1001))
```

**Solution**

* Maximum Number of steps = Worst Case Scenario
* log_2(128) = **7 steps**
* 128->64->32->16->8->4->2->1 = **7 steps**
  

1.2) Suppose you double the size of the list. What’s the maximum number of steps now?

**Solution**

* 256->128->64->32->16->8->4->2->1 = **8 steps**

**The traveling salesperson**

The salesperson has to go to five cities.





He adds up the total distance and then picks the path with the lowest distance. There are 120 permutations with 5 cities, so it will take 120 operations to solve the problem for 5 cities. For 6 cities, it will take 720 operations (there are 720 permutations). For 7 cities, it will take 5,040 operations!



This is a terrible algorithm! The sales person should use a different one, right? But he can’t. This is one of the unsolved problems in computer science. There’s no fast known algorithm for it, and smart people think it’s impossible to have a smart algorithm for this problem. The best we can 
do is come up with an approximate solution.
